```python
from matplotlib import pyplot as plt
import numpy as np

# Generisk iterativ fikspunktmetode.
def fikspunkt(f, x, toleranse = 1.0E-10, N = 500):
    fx = f(x)
    n = 0
    punkter = [(x, fx)] 
    while abs(fx - x) >= toleranse and n < N:
        x = f(x)
        n += 1
        fx = f(x)
        punkter.append((x, fx))
    if n >= N:
        return "Fant ikke fikspunkt for gitt startverdi"
    else: 
        return fx, punkter

# Fikspunktfunksjon.
def f(x):
    return .5 * (x + (2/x))

# Finne fikspunkt.
svar, punkter = fikspunkt(f, .22)

# Skrive ut punkter.
for punkt in punkter:
    print('x = {:1.15f}\t f(x) = {:1.15f}'.format(*punkt))
```

```python
# Lag mesh for grafer.
x = np.arange(0.2, 5, 0.01)

# Plotte f(x) i blått og g(x) = x i rødt.
# Fikspunktet inntreffer når disse skjærer hverandre.
plt.plot(x, f(x), 'b')
plt.plot(x, x, 'r')

# Visualiser fikspunktiterasjonen i grønt.
for x, y in punkter:
    plt.plot([x, x], [x, y], 'g')
    plt.plot([x, y], [y, y], 'g')
plt.show()
```
